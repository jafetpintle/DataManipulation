# Activity 2

In this activity, we will create a line plot to analyze month-to-month trends for items 
sold in the months January through June. The trend will be positive and linear, and will 
be represented using a dotted, blue line, with star markers. The x-axis will be labeled 
**'Month'** and the y-axis will be labeled **'Items Sold'**. The title will say 'Items Sold has been 
Increasing Linearly:'

1. Create a list of six strings for x containing the months January through June.

```python
import numpy as np
x = np.array(['January', 'Frebruary', 'March', 'April', 'May', 'June']) 
```

2. Create a list of six values for y containing values for **'Items Sold'** that start at 1000 
and increase by 200 in each value, so the final value is 2000.

```python
items_sold = np.arange(1000,2001,200)
items_sold
```

3. Generate the described plot.
Check out the following screenshot for the resultant output:
image.png

```python
import matplotlib.pyplot as plt
plt.plot(x, items_sold, '*:b')
plt.title('Items sold has been increasing lineary')
plt.ylabel('Items sold')
plt.xlabel('Month ')
```

#Activity 3: Bar Plot
In this activity, we will be creating a bar plot comparing the number of NBA 
championships among the five franchises with the most titles. The plot will be sorted so 
that the franchise with the greatest number of titles is on the left and the franchise with 
the least is on the right. The bars will be red, the x-axis will be titled **'NBA Franchises'**, 
the y-axis will be titled **'Number of Championships'**, and the title will be programmatic, 
explaining which franchise has the most titles and how many they have. Before 
working on this activity, make sure to research the required NBA franchise data online. 
Additionally, we will rotate the x tick labels 45 degrees using **plt.xticks(rotation=45)**
so that they do not overlap, and we will save our plot to the current directory:

1. Create a list of five strings for x containing the names of the NBA franchises with 
the most titles

```python
x = ['Boston Celtics', 'Minneapolis', 'Chicago Bulls','Philadelphia', 'San Antonio']
```

2. Create a list of five values for y containing values for 'Titles Won' that correspond 
with the strings in x.

```python
y = [17,16,6,6,5]
```

3. Place x and y into a data frame with the column names 'Team' and 'Titles', 
respectively

```python
import pandas as pd

df = pd.DataFrame({'Team':x, 
                   'Titles':y})
df
```

4. Sort the data frame in descending order by 'Titles'.

```python
df = df.sort_values(by='Titles')
df
```

5. Make a programmatic title and save it as title.

```python
team_with_most_titles = df.iloc[len(df)-1]['Team']
title = '{} is the team with the most titles'.format(team_with_most_titles)
```

6. Generate the described plot.


```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
plt.bar(df['Team'], df['Titles'])
plt.title(title)
plt.xlabel('Team')
plt.ylabel('Titles')
plt.xticks(rotation=45)
plt.show()
```

Activity 4: Multiple Plot Types using Subplots

We have learned uptil now how to build, customize, and program line plots, bar plots, 
histograms, scatterplots, and box-and-whisker plots using the functional approach. In 
exercise 19, we were introduced to the object-oriented approach, and in exercise 20, we 
learned how to create a figure with multiple plots using subplots. Thus, in this activity, 
we will be leveraging subplots to create a figure with multiple plots and plot types. We 
will be creating a figure with six subplots.

Datasets:
https://github.com/TrainingByPackt/Data-Science-with-Python/blob/master/Chapter02/Items_Sold_by_Week.csv

https://github.com/TrainingByPackt/Data-Science-with-Python/blob/master/Chapter02/Weight_by_Height.csv

1. Import Items_Sold_by_Week.csv and Weight_by_Height.csv from GitHub and generate a normally distributed array of numbers.

```python
import pandas as pd
import numpy as np

items_data = 'https://github.com/TrainingByPackt/Data-Science-with-Python/blob/master/Chapter02/Items_Sold_by_Week.csv?raw=true'
wbh_data = 'https://github.com/TrainingByPackt/Data-Science-with-Python/blob/master/Chapter02/Weight_by_Height.csv?raw=true'
items_sold_by_week_df= pd.read_csv(items_data, header=0)
weight_by_height_df = pd.read_csv(wbh_data, header=0)
x = np.random.normal(loc=0.5,scale=0.1,size=100)
```

```python
print(items_sold_by_week_df.shape)
print(items_sold_by_week_df.dtypes)
items_sold_by_week_df
```

```python
print(weight_by_height_df.shape)
weight_by_height_df
```

2. Generate a figure with six empty subplots using three rows and two columns that 
do not overlap.

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(nrows=3 , ncols=2)
axes[0,0].set_title('Line')
axes[0,1].set_title('Bar')
axes[1,0].set_title('Horizontal Bar')
axes[1,1].set_title('Histogram')
axes[2,0].set_title('Scatter')
axes[2,1].set_title('Box-and-Whisker')
```

3.  Set the plot titles with six subplots organized in three rows and two columns such 
that do not overlap.

```python
fig, axes = plt.subplots(nrows=3 , ncols=2)
axes[0,0].set_title('Line')
axes[0,1].set_title('Bar')
axes[1,0].set_title('Horizontal Bar')
axes[1,1].set_title('Histogram')
axes[2,0].set_title('Scatter')
axes[2,1].set_title('Box-and-Whisker')
plt.tight_layout()
fig.show()
```

4. On the 'Line', 'Bar' and 'Horizontal Bar' axes, plot 'Items_Sold' by 'Week' from 'Items_
Sold_by_Week.csv'

```python
fig, axes = plt.subplots(nrows=3 , ncols=2)
axes[0,0].set_title('Line')
axes[0,0].plot(items_sold_by_week_df['Week'], items_sold_by_week_df['Items_Sold'])
axes[0,1].set_title('Bar')
axes[0,1].bar(items_sold_by_week_df['Week'], items_sold_by_week_df['Items_Sold'])
axes[1,0].set_title('Horizontal Bar')
axes[1,0].barh(items_sold_by_week_df['Week'], items_sold_by_week_df['Items_Sold'])
axes[1,1].set_title('Histogram')
axes[2,0].set_title('Scatter')
axes[2,1].set_title('Box-and-Whisker')
plt.tight_layout()
fig.show()
```

5. In the 'Histogram' and 'Box-and-Whisker' axes, plot the array of 100 normally 
distributed numbers

```python
fig, axes = plt.subplots(nrows=3 , ncols=2)

axes[0,0].set_title('Line')
axes[0,0].plot(items_sold_by_week_df['Week'], items_sold_by_week_df['Items_Sold'])

axes[0,1].set_title('Bar')
axes[0,1].bar(items_sold_by_week_df['Week'], items_sold_by_week_df['Items_Sold'])

axes[1,0].set_title('Horizontal Bar')
axes[1,0].barh(items_sold_by_week_df['Week'], items_sold_by_week_df['Items_Sold'])

axes[1,1].set_title('Histogram')
axes[1,1].hist(x)

axes[2,0].set_title('Scatter')


axes[2,1].set_title('Box-and-Whisker')
axes[2,0].boxplot(x)

plt.tight_layout()
fig.show()
```

6. In the 'Scatter' axis, plot weight by height with 'Weight_by_Height.csv'.

```python
fig, axes = plt.subplots(nrows=3 , ncols=2)

axes[0,0].set_title('Line')
axes[0,0].plot(items_sold_by_week_df['Week'], items_sold_by_week_df['Items_Sold'])

axes[0,1].set_title('Bar')
axes[0,1].bar(items_sold_by_week_df['Week'], items_sold_by_week_df['Items_Sold'])

axes[1,0].set_title('Horizontal Bar')
axes[1,0].barh(items_sold_by_week_df['Week'], items_sold_by_week_df['Items_Sold'])

axes[1,1].set_title('Histogram')
axes[1,1].hist(x)

axes[2,0].set_title('Scatter')
axes[2,0].scatter(weight_by_height_df['Weight'],weight_by_height_df['Height'])


axes[2,1].set_title('Box-and-Whisker')
axes[2,1].boxplot(x)

plt.tight_layout()
fig.show()
```

7. Label the x- and y-axis in each subplot.

```python
fig, axes = plt.subplots(nrows=3 , ncols=2)

axes[0,0].set_title('Line')
axes[0,0].plot(items_sold_by_week_df['Week'], items_sold_by_week_df['Items_Sold'])
axes[0,0].set_xlabel('Week')
axes[0,0].set_ylabel('Items Sold')

axes[0,1].set_title('Bar')
axes[0,1].bar(items_sold_by_week_df['Week'], items_sold_by_week_df['Items_Sold'])
axes[0,1].set_xlabel('Week')
axes[0,1].set_ylabel('Items Sold')

axes[1,0].set_title('Horizontal Bar')
axes[1,0].barh(items_sold_by_week_df['Week'], items_sold_by_week_df['Items_Sold'])
axes[1,0].set_xlabel('Week')
axes[1,0].set_ylabel('Items Sold')

axes[1,1].set_title('Histogram')
axes[1,1].hist(x)
axes[1,1].set_xlabel('Normally Distributed Numbers')

axes[2,0].set_title('Scatter')
axes[2,0].scatter(weight_by_height_df['Weight'],weight_by_height_df['Height'])
axes[2,0].set_xlabel('Weight')
axes[2,0].set_ylabel('Height')

axes[2,1].set_title('Box-and-Whisker')
axes[2,1].boxplot(x)
axes[2,1].set_xlabel('Normally Distributed Numbers')

plt.tight_layout()
fig.show()
```
