#Exercise 13: Functional Approach - Line Plot


```python
#Generating an array of numbers for the horizontal azis rangin from 0 to 10 in 20 evenly spaced values
import numpy as np

x = np.linspace(0,10,20)
x
```

```python
y = x**3
```

```python
#Ploting
import matplotlib.pyplot as plt

plt.plot(x,y, 'D-k')
plt.xlabel('Lnearly Spaced Numbers')
plt.ylabel('y values')
plt.title('x by x cubed', fontsize=22)
plt.show()
```

# Exercise 14: Functional Approach - Add a second Line to the Line Plot

```python
y2 = x**2
```

```python
#Ploting
import matplotlib.pyplot as plt
#First Excersise
plt.figure(figsize=(10,5))
plt.plot(x,y, 'D-k', label='x cubed')
plt.plot(x,y2, '--r', label='x squared')
plt.xlabel('Lnearly Spaced Numbers')
plt.ylabel('y values')
plt.title('As x increases, \nx cubed (black) increases \nat a Greter Rate than \nx Squared(red)', fontsize=22)
plt.legend(loc='upper left')
plt.show()
```

# Exercise 15: Creating a Bar Plot

```python
#Crating a list of items types
x = ['Shirts', 'Pants', 'Shorts', 'Shoes']
```

```python
y = [1000,1200,800,1800]
```

```python
import matplotlib.pyplot as plt

plt.bar(x,y)
plt.title('Sales Revenue by Item Type')
plt.xlabel('Item Type')
plt.ylabel('Sales Revenue ($)')
plt.show()
```

```python
#Making a plot that changes tittle
index_of_max_y = y.index(max(y))
```

```python
#Saving the most sold item
most_sold_item = x[index_of_max_y]
```

```python
#Makin a tittle programming
plt.barh(x,y)

plt.title('{} Produce the Most Sales Revenue'.format(most_sold_item))

plt.ylabel('Item Type')
plt.xlabel('Sales Revenue ($)')
plt.show()
```
